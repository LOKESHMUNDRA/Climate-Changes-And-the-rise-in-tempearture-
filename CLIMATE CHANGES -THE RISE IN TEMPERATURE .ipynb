### Global land temperature state wise

```python
df_states = pd.read_csv("GlobalLandTemperaturesByState.csv")
df_states.head(10)

```

```python
df_states.tail(10)
```

```python
df_states.isna().sum()
```

```python
states = df_states[df_states['Country']=='India']
states = states.groupby('State')['AverageTemperature'].mean().reset_index()
states.sort_values('AverageTemperature', inplace = True)
states = states[:10]
states.plot(kind = 'bar', x = 'State', y = 'AverageTemperature' ,title  = 'Top 10 Coolest City', color = 'deepskyblue')
```

```python
states = df_states[df_states['Country']=='India']
states = states.groupby('State')['AverageTemperature'].mean().reset_index()
states.sort_values('AverageTemperature', inplace = True,ascending = False)
states = states[:10]
states.plot(kind = 'bar', x = 'State', y = 'AverageTemperature' ,title  = 'Top 10 Hotest City', color = 'deepskyblue')
```
